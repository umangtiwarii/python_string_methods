# strint.isidentifier()

```python
string = "This_is_python_class_10"
print(string.isidentifier())

string = "This is python class 10"
print(string.isidentifier())


# A string is a valid identifier if:
# ✅ It contains only:
# Letters (a–z, A–Z)
# Digits (0–9)
# Underscore _
# ✅ It does NOT start with a digit
# ❌ It does NOT contain spaces
# ❌ It does NOT contain special characters (@ # $ % etc.)
# ❌ It is not a Python keyword (like if, for, class, etc.
```
